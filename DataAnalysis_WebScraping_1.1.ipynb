```python
import requests
import time
import datetime
import csv
import smtplib

from bs4 import BeautifulSoup

isHeaderIncludedInCSV = False

def getURLHeaders():
    #http://httpbin.org/get
    urlHeaders = { 
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/96.0.4664.45 Safari/537.36",
        "Accept-Encoding": "gzip, deflate", 
        "Accept-Language": "en-US,en;q=0.9,bn;q=0.8",
        "Accept": "text/html,application/xhtml+xml,application/xml;q=0.9,image/avif,image/webp,image/apng,*/*;q=0.8,application/signed-exchange;v=b3;q=0.9",
        "Upgrade-Insecure-Requests": "1", 
    }
    return urlHeaders

def getURL():
    URL = 'https://www.amazon.com/Harry-Potter-Gryffindor-Quidditch-T-Shirt/dp/B00XIMWA3M/ref=mp_s_a_1_12?_encoding=UTF8&c=ts&keywords=Men%27s+Novelty+T-Shirts&qid=1637258467&qsid=132-3137104-7293019&s=apparel&sr=1-12&sres=B071FWLYKQ%2CB01LB72TBS%2CB00KRIR134%2CB07Q8FK5NS%2CB071GWH53T%2CB00XJOZN92%2CB005CSUWW6%2CB007R1OJPM%2CB0778TLRZV%2CB00XIMWA3M%2CB01IIHM20W%2CB06XYRRK7X%2CB00MNTVCAS%2CB01L1BMJ2S%2CB07N12HR1D%2CB06Y33MW6J%2CB00XLESTXC%2CB07TMQ82NK%2CB00L2RN3ZO%2CB01LB70K50&srpt=SHIRT&ts_id=9056987011'
    return URL

def getProductSoup():
    req = requests.get(getURL(),headers = getURLHeaders() )
    productSoup = BeautifulSoup(req.content, 'html.parser')
    return productSoup

def readHeaderFromCSV(csvFileName):
    rows = []
    with open(csvFileName, 'r') as file:
        csvreader = csv.reader(file)
        header = next(csvreader)
#         for row in csvreader:
#             rows.append(row)
    return header

def isAppendIntoCSV(csvFileName):
    header = readHeaderFromCSV(csvFileName)
    if(len(header) == 0):
        isHeaderIncludedInCSV = False
    else:
        isHeaderIncludedInCSV = True
    return isHeaderIncludedInCSV
    
def writeIntoCSV(csvFileName,csvHeader,data):
    #csvHeader = ['title','price','ratings','productDescription','customerReview','date']
    with open(csvFileName, 'w', newline='', encoding='UTF8') as csvfile:
        writer = csv.writer(csvfile)
        writer.writerow(csvHeader)
        writer.writerow(data)

def appendIntoCSV(csvFileName,data):
    with open(csvFileName, 'a+', newline='', encoding='UTF8') as csvfile:
        writer = csv.writer(csvfile)
        writer.writerow(data)

def getProductInfo(soup):
    # get web scaped content.
    
    #set csv file name.
    csvFileName = r'G:\source\DS\PortfolioProject\DataAnalysis_WebScraping\names.csv'
    
    # prepare data header.
    csvHeader = ['title','price','ratings','productDescription','customerReview','date']
    
    # prepare data .
    title = soup.find(id="productTitle").get_text().strip()
    
    price = '';
    try:
        for tag in soup.find_all(class_="a-price a-text-price a-size-medium apexPriceToPay"):
            if(tag.next_element.name=='span'):
                price = tag.next_element.text[1:].strip()
                price = float(price)
    except:
        price = ''
    
    ratings = soup.select('.a-icon-alt')[0].get_text().split(' ')[0].strip()
    ratings = float(ratings)
    
    productDescription = soup.find("div", {"id": "productDescription"}).get_text().strip()
    
    customerReview = soup.select('#acrCustomerReviewText')[0].get_text().split(' ')[0].strip().replace(",", "")
    customerReview = int(customerReview)
    
    today = datetime.date.today()
    #print(today)
    
    data = [title,price,ratings,productDescription,customerReview,today]

    # write / append into csv
              
    isAppend = isAppendIntoCSV(csvFileName)
    
    if(isAppend):
        appendIntoCSV(csvFileName,data)
    else:
        writeIntoCSV(csvFileName,csvHeader,data)              
    
```

```python
counter = 0
maxLimit = 1000
timerSleepInSeconds = 5

print('-------------------Start----------------------')
soup = getProductSoup()

while(counter < maxLimit ):
    getProductInfo(soup)
    #print(counter)
    #time.sleep(5)
    counter += 1 
print('-------------------Done----------------------')
```
